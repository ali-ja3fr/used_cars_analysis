```python
cars_df['MSRP_formatted']= cars_df['MSRP'].apply(lambda x: "${:,.0f}".format(x))
cars_df
```

```python
avg_popularity_sedan = cars_df[cars_df['Vehicle Style'] == 'Sedan']['Popularity'].mean()
print(f"Average Popularity for Sedans: {avg_popularity_sedan:.2f}")
```

```python
avg_popularity_suv = cars_df[cars_df['Vehicle Style'].str.contains('suv', case=False, na=False)]['Popularity'].mean()
print(f"Average Popularity for SUVs: {avg_popularity_suv:.2f}")
```

```python
avg_popularity_by_style = cars_df.groupby('Vehicle Style')['Popularity'].mean().sort_values(ascending=False)

print(avg_popularity_by_style)
```

```python
average_year = cars_df['Year'].mean()
print(f"Average Year of Cars: {average_year:.2f}")
```

```python
print(f"Average Year of Cars (Rounded): {round(average_year)}")
```

```python
newer_cars = cars_df[cars_df['Year'] > 2015]

print(f"Number of cars from after 2015: {len(newer_cars)}")
newer_cars.sample(20) 
```

```python
# Ensure MSRP is numeric (if not already)
cars_df['MSRP_numeric'] = cars_df['MSRP_formatted'].replace('[\$,]', '', regex=True).astype(float)

# Filter for the 3 specific vehicle styles
target_styles = ['Cargo Van', '2dr SUV', 'Passenger Van']
filtered_df = cars_df[cars_df['Vehicle Style'].isin(target_styles)]

# Calculate the average MSRP (numeric)
avg_msrp_raw = filtered_df.groupby('Vehicle Style')['MSRP_numeric'].mean()

# Display the result
print(avg_msrp_raw)
```

```python
# Filter only Cargo Vans
cargo_vans = cars_df[cars_df['Vehicle Style'] == 'Passenger Van']

# Group by Make and count
cargo_van_counts = cargo_vans['Make'].value_counts()

# Display results
print(cargo_van_counts)
```
